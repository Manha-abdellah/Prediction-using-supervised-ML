**Predict the percentage of a student based on # of study hours**

1. Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

2. Importing Data

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported successfully")
```

```python
data.head()
```

3. Plotting Data

**Plot the graph for analysis of the dataset**

```python
data.plot(x="Hours", y="Scores", style='1')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Scores')
plt.show()
```

Data Preparation for the Model

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,1].values
```

4. Modeling and Training Data

```python
#dividing data for triing and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

#Training the Algorithm

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

# from sklearn.ensemble import RandomFOrestRegressor
# regressor = RandomForestRegressor(n_estimators = 1000, random_state = 42)

regressor.fit(x_train,y_train)
print("Training Complete")
```

5. Testing Data

```python
print(x_test)
print("Prediction of Score")
y_pred = regressor.predict(x_test)
print(y_pred)
```

6. Predicting accuracy of Model

```python
df = pd.DataFrame({'Actual': y_test,'Predicted': y_pred})
df
```

7.Prediction with Custom Input


```python
hours = [[9.25]]
pred = regressor.predict(hours)
print(pred)
```

7.Evaluating the Model

```python
from sklearn import metrics
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
```
